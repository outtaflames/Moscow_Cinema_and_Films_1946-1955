```python
class MoscowFilms(): # создаем класс для общих файлов по фильмам в Москве
  def __init__(self, filename: str, Cinema, First_day_of_screening, Screening_Days, Title_original):
    self.filename = filename

  def read_file(self):
    with open (self.filename, 'r') as myfile:
      mfilms = csv.reader(myfile)
      return mfilms
```
